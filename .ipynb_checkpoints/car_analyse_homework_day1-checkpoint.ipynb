# Action3: 对汽车质量数据进行统计


数据集：car_complain.csv
600条汽车质量投诉
Step1，数据加载
Step2，数据预处理
拆分problem类型 => 多个字段
Step3，数据统计
对数据进行探索：品牌投诉总数，车型投诉总数
哪个品牌的平均车型投诉最多


```python
#加载数据
import pandas as pd
result = pd.read_csv('car_complain.csv')
```

```python
#将抱怨列按照逗号打散
result=result.drop('problem', 1).join(result['problem'].str.split(',',expand=True))
```

```python
#采用逆透视的方法，把二维数据一维化
t1=pd.melt(result,id_vars=['id','brand','car_model','type','desc','datetime','status']).drop('variable', 1)
#这里进行了一个奇怪的保存，再读入。因为直接对t1去除空格，会导致t1里同时存在''及None，
#None不会被删除。发现保存后读取能解决这个问题
t1.to_csv('test.csv',encoding='gbk')
t2=pd.read_csv('test.csv',encoding='gbk')
t2.dropna(axis=0, how='any',inplace=True)
```

```python
#投诉计数函数，传入需要分组的列表，比如['brand','car_model'],最后输出投诉计数和投诉内容总数
def claimCount(by):
    #品牌投诉计数
    s1=t2.groupby(by)['id'].nunique().sort_values(ascending=False).rename('投诉计数')
    #品牌投诉总数
    s2=t2.groupby(by)['id'].count().sort_values(ascending=False).rename('投诉内容计数')
    return pd.DataFrame([s1,s2]).T
```

```python
claimCount(['brand'])
```

```python
claimCount(['brand','car_model'])
```

```python
#计算平均抱怨数
avgClaims=claimCount(['brand']).div(t2.groupby('brand')['car_model'].nunique(),axis=0)
```

```python
#更改列名，并排序
avgClaims.columns=['品牌按车型投诉平均数', '品牌按车型投诉内容平均数']
avgClaims.sort_values(by='品牌按车型投诉平均数',ascending=False)
```
